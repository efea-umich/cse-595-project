```python
import pandas as pd
import yaml
import lightning as L

import torch

from enhance_data import DataEnhancer
from basketball_config import BasketballConfig
from basketball_dataset import BasketballDataset
from time_series_transformer_model import TimeSeriesTransformerModel
from torch.utils.data import DataLoader

from sanity_check_model import SanityCheckTSModel
```

```python
with open("play-types.yaml", 'r') as f:
    config = BasketballConfig.model_validate(yaml.safe_load(f))
```

```python
df = pd.read_csv("data/NBA_PBP_2015-16.csv")
```

```python
df = DataEnhancer(df, config).enhance_data(df)
```

```python
df.head()
```

```python
pd.get_dummies(df[["PlayType", "TurnoverType"]])
```

```python
basketball_dataset = BasketballDataset(df, config, num_sequences=20000)
```

```python
train, val = torch.utils.data.random_split(basketball_dataset, [0.8, 0.2])
```

```python
example_x, example_y = basketball_dataset[0]
print(f"Loaded example data with sequence length {example_x.shape[0]} and features per step {example_x.shape[1]}")
```

```python
train_dataloader, val_dataloader = DataLoader(train, batch_size=32, shuffle=True), DataLoader(val, batch_size=32, shuffle=True)
```

```python
model = TimeSeriesTransformerModel(
    seq_len=example_x.shape[0], features_per_step=example_x.shape[1], embed_dim=16, n_heads=2
)
sanity_check_model = SanityCheckTSModel(
    seq_len=example_x.shape[0], features_per_step=example_x.shape[1]
)
```

```python
trainer = L.Trainer(max_epochs=10)
```

```python
trainer.fit(model, train_dataloader, val_dataloaders=[val_dataloader])
```
